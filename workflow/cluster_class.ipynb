# Check Molecule's Methos in Cluster Object

```python
w1 = [("O", -2.5, 0, 0), ("H", -2.5 + 0.58708, 0.75754, 0), ("H", -0.58708 - 2.5, 0.75754, 0)]
w2 = [("O", 0, 0, 0), ("H", 0.58708, 0.75754, 0), ("H", -0.58708, 0.75754, 0)]
w3 = [("O", 2.5, 0, 0), ("H", 2.5 + 0.58708, 0.75754, 0), ("H", -0.58708 + 2.5, 0.75754, 0)]
trimer = Cluster(w1,w3)
trimer += w2
print(f"Type of class: {type(trimer)}")
print(f"Number of Atoms: {trimer.GetNumAtoms()}")
print(f"Atomic symbols: {trimer.GetAtomicSymbols()}")
print(f"Atomic Number: {trimer.GetAtomicNumbers()}")
print(f"Atomic coordinates:\n {trimer.GetAtomicCoordinates()}")
print(f"Atomic masses: {trimer.GetAtomicMasses()}")
print(f"Molecule mass: {trimer.GetMolMass()}")
print(f"Molecule charge: {trimer.GetMolCharge()}")
print(f"Molecule mutiplicity: {trimer.GetMolMultiplicity()}")
print(f"Print Molecule dictionary:\n {trimer.GetMolDict()}")
print(f"Print Molecule list:\n {trimer.GetMolList()}")
print(f"Center of mass of molecule: {trimer.GetMolCM()}")
print(f"Axes principal of molecule: {trimer.GetMolPrincipalAxes()}")
print(f"Print Atoms for index: {trimer.GetAtomWithIndix(0)}")
print(trimer)
trimer
```

```python
print("Cluster Dict:\n ",trimer.GetClusterDict())
print("Molecule Dict:\n ",trimer.GetMolDict())
print("Cluster List:\n ",trimer.GetClusterList())
print("Molecule List:\n ",trimer.GetMolList())
print("Block XYZ from Molecule Class Method:\n ",trimer.GetBlockXYZ())
```
