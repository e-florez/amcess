# The Atomic and Molecular Energy Surface Sampler (AMCESS) [1]

AMCESS is an end-to-end package implemented in Python3.9 to sample the PES and generate candidate structures for critical points.

 - The AMCESS package uses simple input files and automates common procedures.

 - The package also allows the user to perform local searches around defined regions.

 - The PES is generated computing the electronic energy using standard and powerful quantum chemistry packages such as **PySCF** [2] and **Psi4** [3], also implemented in Python.

 - The AMCESS main purpose is to be a user friendly package, easy to install, import, and run, available in most platforms and open-source. 
 
 - As a Python module, AMCESS can be integrated into any workflow.

# ASCEC

A previous version of AMCESS, called ASCEC [1] (spanish acronym Annealing Simulado con Energía Cuántica) was written in FORTRAN77 and was successfully used in the wide range of research and academic applications. 


- From atomic cluster and molecular cluster, the ASCEC package has produced novel results (structure never seen before) in more than 30 papers published in the literature.

# 3.2 Dual Annealing

This method follows the Generalized Simulated
Annealing, a generalization of the Simulated Annealing of the ASCEC algorithm.

Instead of a random move in the molecule, the move follows the Tsallis-Stariolo form of the Cauchy-Lorentz visiting distribution. Wich also change the rate of cooling $T$. Moreover, it uses a generalized Metropolis algorithm for the acceptance algorithm. 

Since GSA has a nonlocal search and has the probability of long jumps even at a low temperature, it has a large probability of escaping from a local minimumCaptura de pantalla de 2021-12-21 19-55-59.png

#3.3 Simplicial Homology Global Optimization (SHGO)

Global optimization algorithm based on application of simplicial integral homology and combinatorial topology, implemented into the Scipy Library.
This approach is used to confine regions with local minima. Then, a local minimization routine is used to characterize the total energy of the cluster.

Captura de pantalla de 2021-12-21 21-00-58.png

This method allow us to work with constrained functions, and gives competitive results compared to others global optimization methods (Endres et al. 2018)
Captura de pantalla de 2021-12-21 21-24-55.png 


The variables are the distance between the center of mass of the molecules that compound the cluster, and the Euler’s angles between them.

## System object:

- Atom
- Molecule
- Cluster


```python
    from amcess.base_molecule import Molecule, Cluster

    water_dict = {
        "atoms": [("O", 0, 0, 0), ("H", 0.587, 0.757, 0), ("H", -0.587, 0.757, 0)],
        "charge": 0,
        "multiplicity": 1
    }
    water_mol = Molecule.from_dict(water_dict)

    water_pentamer = Cluster(5 * water_mol)
    water_pentamer.sphere_radius = 2.0
    water_pentamer.sphere_center = (0.0, 0.0, 1.0)
```







## Search Engine
- SearchConfig

## Electronic Energy
- ElectronicEnergy

```python
from amcess.search_engine import SearchConfig

obj_bayesian = SearchConfig(water_pentamer, search_methodology="Bayesian")
obj_bayesian.run(initer=10, maxiter=10, seed=123)

```
